# Derivadas
* Tensorflow puede calcular derivadas y gradientes.
* Los gradientes se calculan usando la regla de la cadena de manera recursiva:

$$ {\partial f(g(x)) \over \partial x} = {\partial f(g(x)) \over \partial g(x)}\cdot {\partial g(x) \over \partial x} $$

Puede derivar cualquier funci√≥n siempre que pueda derivar las operaciones elementales que usa.


```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline

###Crea una sesi√≥n 
sess = tf.Session()

x = tf.placeholder('float32')

f = 3*x


derivada = tf.gradients(f, x)[0]

x_array = np.linspace(-3,3)
#print(x_array)
f_, f_derivada = sess.run([f,derivada],
                                 {x:x_array})

plt.plot(x_array, f_, label="f")
plt.plot(x_array, f_derivada, label="df")
plt.legend();
```
